```python
import pandas as pd
source = 'LSE_DA201_Week_3_files/LSE_DA201_Week_3_files/Data/'
movies = pd.read_excel(source + 'movies_merge.xlsx')
ott = pd.read_csv(source + 'ott_merge.csv')
```

```python
movies.head()
```

```python
merged = movies.merge(ott, how='left', on='ID')
concated = pd.concat([movies, ott], axis=0)
appended = movies.append(ott)
```
